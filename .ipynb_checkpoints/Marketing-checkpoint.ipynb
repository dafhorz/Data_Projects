# Cleaning and Transforming Data

## Relabeling 'Marital_Status', Renaming 'Income' and Dealing with Null Values

Seems like the *Income* column has been miswritten. Let's correct it.

```python
df.columns
```

```python
df.rename(columns={' Income ':'Income'}, inplace=True)
```

Since the Campaigns are not in ascending order, lets reorder the columns.

```python
df = df[['ID', 'Year_Birth', 'Education', 'Marital_Status', 'Income', 'Kidhome',
       'Teenhome', 'Dt_Customer', 'Recency', 'MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
       'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4',
       'AcceptedCmp5', 'Response', 'Complain', 'Country']]
df.head()
```

Now, we should relabel *YOLO*, *Alone* and *Absurd* as *Single*.

```python
df['Marital_Status'].unique()
```

```python
df['Marital_Status'].replace({'YOLO':'Single', 'Alone':'Single', 'Absurd':'Single'}, inplace=True)
df['Marital_Status'].unique()
```

```python
df.isnull().sum()
```

```python
df.query('Income != Income')
```

```python
df.isnull().sum() * 100 / len(df)
```

Even though there's only a $1\%$ ratio of missing values in the *Income* column, let's fill each with the column's median. <br>
Observe that there are a few really high values compared to the rest.

```python
df['Income']
max(df['Income'])
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
```

## Changing dates to Timestamp

```python
df.head()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format = '%Y-%m-%d').dt.date #The last .dt.date just keeps the date, without the hour.
df.info()
```

## Dropping Outliers: IQR

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

Of course, several values don't make sense right here, so let's remove them.

```python
df.describe()
```

These are the outliers we are about to drop. <br> <br>
Two other probable values to consider: "Recency", "Year Birth".

```python
df[ (  (df['Income'] < Q1['Income'] - 1.5*IQR['Income']) | (1.5* IQR['Income'] + Q3['Income'] < df['Income'])  ) ]
```

```python
df = df[ ~(  (df['Income'] < Q1['Income'] - 1.5*IQR['Income']) | (1.5* IQR['Income'] + Q3['Income'] < df['Income'])  ) ]
df=df.reset_index(drop = True)
```

```python
ax = df.Year_Birth.hist(figsize=(10, 5), bins = 30)
ax.set_ylim(top = 10)
```

```python
df = df[ df['Year_Birth'] > 1920 ]
```

```python
df.shape
```

## Exporting CSV: noNull & No Obvious Outliers

```python
df.to_csv('processed/data_noNull_noObviousOutliers.csv')
```

# Load to MySQL

```python
import pymysql as sql
```

```python
user = 'root'
passwd = '----------'

con = sql.connect(host = '127.0.0.1', user = user, passwd = passwd)
cursor = con.cursor()

```

```python
cursor.execute("USE MarketingCampaign;")
cursor.execute("DROP TABLE IF EXISTS Campaign;")
cursor.execute("""
CREATE TABLE IF NOT EXISTS Campaign(
    ID INT NOT NULL,
    Year_Birth INT NOT NULL,
    Education VARCHAR(30) NOT NULL,
    Marital_Status VARCHAR(30) NOT NULL,
    Income FLOAT NOT NULL,
    Kidhome INT NOT NULL,
    Teenhome INT NOT NULL,
    Dt_Customer DATE NOT NULL,
    Recency INT NOT NULL,
    MntWines INT NOT NULL,
    MntFruits INT NOT NULL,
    MntMeatProducts INT NOT NULL,
    MntFishProducts INT NOT NULL,
    MntSweetProducts INT NOT NULL,
    MntGoldProducts INT NOT NULL,
    NumDealsPurchases INT NOT NULL,
    NumWebPurchases INT NOT NULL,
    NumCatalogPurchases INT NOT NULL,
    NumStorePurchases INT NOT NULL,
    NumWebVisitsMonth INT NOT NULL,
    AcceptedCmp1 INT NOT NULL,
    AcceptedCmp2 INT NOT NULL,
    AcceptedCmp3 INT NOT NULL,
    AcceptedCmp4 INT NOT NULL,
    AcceptedCmp5 INT NOT NULL,
    Response INT NOT NULL,
    Complain INT NOT NULL,
    Country VARCHAR(30) NOT NULL,
    PRIMARY KEY (ID) );
""")

con.commit() 
```

```python
valueList = []

colNames = list(df.columns)
for index,row in df.iterrows():
    arr=['"' + str(row[col]) + '"' for col in colNames]
    valueList.append( '(' + ','.join(arr) + ')' )

values = ', '.join(valueList)+';'
```

```python
cursor.execute("""
INSERT INTO Campaign values %s
""" % values)

con.commit()
```
